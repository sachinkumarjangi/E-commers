```python
# Make sure to run the cell with import statements first
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df['Order Date']=pd.to_datetime(df['Order Date'])
df['ship date']=pd.to_datetime(df['ship date'])
df.info()
```

```python
df.head()
```

```python
df['order month']=df['Order Date'].dt.month
df['order year']=df['Order Date'].dt.year
df['order day of weak']=df['Order Date'].dt.dayofweek
df.head()
```

```python
sales_by_month=df.groupby('order month')['Sales'].sum()
plt.figure(figsize=(10,6))
sns.lineplot(x=sales_by_month.index, y=sales_by_month.values, palette='viridis')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Total Sales by Month')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```

```python
sales_by_catagory=df.groupby('Category')['Sales'].sum()
plt.figure(figsize=(10,6))
sns.barplot(x=sales_by_catagory.index, y=sales_by_catagory.values, palette='viridis')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.title('Total Sales by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
sales_by_subcatagory=df.groupby('Sub-Category')['Sales'].sum()
plt.figure(figsize=(10,6))
sns.barplot(x=sales_by_subcatagory.index, y=sales_by_subcatagory.values, palette='viridis')
plt.xlabel('Sub-Category')
plt.ylabel('Total Sales')
plt.title('Total Sales by Sub-Category')
plt.xticks(rotation=45)
plt.show()
```

```python
monthly_profit=df.groupby('order month')['Profit'].sum()
plt.figure(figsize=(10,6))
sns.scatterplot(x=monthly_profit.index, y=monthly_profit.values, palette='viridis')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.title('Total Profit by Month')
plt.show()
```

```python
profit_by_subcatagory=df.groupby('Sub-Category')['Profit'].sum()
plt.figure(figsize=(10,6))
sns.lineplot(x=profit_by_subcatagory.index, y=profit_by_subcatagory.values, palette='viridis')
plt.xlabel('Sub-Category')
plt.ylabel('Total Profit')
plt.title('Total Profit by Sub-Category')
plt.xticks(rotation=45)
plt.show()
```
